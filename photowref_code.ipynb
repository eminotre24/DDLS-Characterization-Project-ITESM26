```python
from PIL import Image
from pathlib import Path

# ---- Reference for pixel/mm ----
reference = Image.open("./foto_001.jpg")
w_ref, h_ref = reference.size

# ---- Data from Camera ----
x_mmsize = 23.5 # mm
y_mmsize = 15.6
length = 163.18
x_og = 43.8 # reales
y_og = 73.34
#x_og = x_mmsize/2
#y_og = y_mmsize/2

x_pixmm = w_ref / x_mmsize # pix/mm
y_pixmm = h_ref / y_mmsize

pixmm = (x_pixmm + y_pixmm)/2 # Mean


def pad_image(img, model, SN, distance, sizex, sizey, ref_x, ref_y):
    """ 
    img: image object
    model, SN, distance: strings, for name
    sizex, sizey: size in mm of the path in each axis (num of grids times distance of the camera in that axis)
    ref_x, ref_y: references in mm
    """
    # Uneven padding in pixels - OG
    #pad_left = int((ref_x) * pixmm)
    #pad_right = int((length + x_mmsize - ref_x - sizex) * pixmm)
    #pad_top = int((ref_y) * pixmm)
    #pad_bottom = int((length + y_mmsize - ref_y - sizey) * pixmm)

    # Invertido x
    #pad_right = int((ref_x) * pixmm)
    #pad_left = int((length + x_mmsize - ref_x - sizex) * pixmm)
    #pad_top = int((ref_y) * pixmm)
    #pad_bottom = int((length + y_mmsize - ref_y - sizey) * pixmm)

    # Con referencias
    #pad_left = int((ref_x + x_og) * pixmm)
    #pad_right = int((length + x_mmsize - x_og - ref_x - sizex) * pixmm)
    #pad_top = int((ref_y - y_og) * pixmm)
    #pad_bottom = int((length + y_mmsize + y_og - ref_y - sizey) * pixmm)

    # Padding Kisfa
    pad_left = int((81.59 + ref_x - x_og - x_mmsize/2) * pixmm)
    pad_right = int((length + x_og - ref_x - sizex - 81.59) * pixmm)
    pad_top = int((81.59 + ref_y - y_og - y_mmsize/2) * pixmm)
    pad_bottom = int((length + y_og - ref_y - sizey - 81.59) * pixmm)

    width, height = img.size

    new_width = width + pad_left + pad_right
    new_height = height + pad_top + pad_bottom

    newImage = Image.new(
        mode=img.mode,
        size=(new_width, new_height),
        color=(0, 0, 0)
    )

    # Paste image shifted by left and top padding
    newImage.paste(img, (pad_left, pad_top))

    folder = Path(f"./DDLS_{model}_SN_{SN}")
    folder.mkdir(parents=True, exist_ok=True)

    newImage.save(
        f"./DDLS_{model}_SN_{SN}/DDLS_{model}_SN_{SN}_{distance}mm_padded.jpg"
    )

    print("image generated with dimensions", new_width, new_height)
```
